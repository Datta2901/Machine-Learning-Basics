```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,36,49]
```

## Line Graph

```python
plt.plot(x,y)
plt.title("y = x * x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.savefig("plot.jpg")
# to save figure
```

## Bar Graphs

```python
plt.bar(x,y,color = "red")
```

```python
plt.barh(x,y,color = "blue")
```

## Scatter Plots

```python
plt.scatter(x,y,alpha=1,s=20)
# intensity alpha
# size s
```
